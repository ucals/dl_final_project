```python
import pandas as pd
import numpy as np
```

```python
%%time
df = pd.read_csv('tweets.csv', sep=';', parse_dates=['timestamp'])
print(len(df))
df.head()
```

```python
dfi = pd.read_csv('influencers.csv', sep=';')
influencers = dfi['influencer'].tolist()
print(influencers)
print(len(influencers))
```

```python
%%time
dff = df[df['user'].isin(influencers)]
dff = dff.sort_values('timestamp').reset_index(drop=True)
```

```python
dff.to_csv('tweets_influencers.csv', index=False)
```
